```python
## LIBRERÍAS
import matplotlib.pyplot as plt
import numpy as np

## DEFINICION DE VARIABLES
x = np.arange(0,100,10)
y = 5.31 * (10 ** -3) * (81 ** 0.5) / x
y2 = 5.31 * (10 ** -3) * (1 ** 0.5) / x

## GRÁFICAS
plt.plot(x, y, 'Black')
plt.plot(x, y2, '--')
plt.title('Figura 4.3')
plt.xlabel('Conductividad (mS/m)')
plt.ylabel('Profundidad pelicular de penetracion, D(m)')
plt.show()
```
